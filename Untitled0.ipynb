```python
import pandas as pd
import numpy as np
import csv
import matplotlib.pyplot as plt
fig = plt.subplots(figsize =(12, 8))
barWidth = 0.25



df = pd.read_csv("/content/sales (2).csv")
w1 = df['week 1 '].tolist()
w2 = df['week 2'].tolist()
w3= df ['week3'].tolist()


br1 = np.arange(len(w1))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]
plt.bar( br1,w1, color ='r', width = barWidth ,linewidth=2, linestyle = 'dashed', 
        edgecolor ='purple', label ='w1')
plt.bar( br2,w2, color ='g', width = barWidth, linewidth=2, linestyle = 'dashed',
        edgecolor ='grey', label ='w2')
plt.bar( br3,w3, color ='b',width = barWidth, linewidth=2, linestyle = 'dashed',
        edgecolor ='grey', label ='w3')
 
# Adding Xticks
plt.xlabel('DAY')
plt.ylabel('SALES(in Rs)')
plt.xticks([r + barWidth for r in range(len(w1))],
        ['MONDAY', 'TUESDAY', 'WEDNESDAY', 'THURSDAY', 'FRIDAY','SATURDAY','SUNDAY'])
 
plt.legend()
plt.show()
```

```python
import sklearn
from sklearn.datasets import load_boston
import pandas as pd
import seaborn as sns
  
# Load the dataset
dat = pd.read_csv('/content/Iris.csv')
  
# IQR ranges
Q1 = np.percentile(dat['SepalWidthCm'], 25,
                interpolation = 'midpoint')
  
Q3 = np.percentile(dat['SepalWidthCm'], 75,
                interpolation = 'midpoint')
IQR = Q3 - Q1
  
print("Old Shape: ", dat.shape)
  
# Upper bound
upper = np.where(dat['SepalWidthCm'] >= (Q3+1.5*IQR))
  
# Lower bound
lower = np.where(dat['SepalWidthCm'] <= (Q1-1.5*IQR))
  
# Removing the Outliers
dat.drop(upper[0], inplace = True)
dat.drop(lower[0], inplace = True)
  
print("New Shape: ", dat.shape)
  
sns.boxplot(x='SepalWidthCm', data=dat)
```
